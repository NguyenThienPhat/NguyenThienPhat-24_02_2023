```python
from skfuzzy.control.controlsystem import ControlSystemSimulation
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
from skfuzzy import control as ctrl
setpoint= ctrl.Antecedent(np.arange(18,30,1),'setpoint')
sensor = ctrl.Antecedent(np.arange(-5,60,1),'sensor')
pwm = ctrl.Consequent(np.arange(0,225,1),'pwm')


setpoint['low']=fuzz.trimf(setpoint.universe,[18,18,24])
setpoint['medium']=fuzz.trimf(setpoint.universe,[18,24,30])
setpoint['high']=fuzz.trimf(setpoint.universe,[24,30,30])


sensor['vlow']=fuzz.trimf(sensor.universe,[-5,-5,35])
sensor['low']=fuzz.trimf(sensor.universe,[-5,17,35])
sensor['medium']=fuzz.trimf(sensor.universe,[17,35,48])
sensor['high']=fuzz.trimf(sensor.universe,[35,48,60])
sensor['vhigh']=fuzz.trimf(sensor.universe,[35,60,60])

pwm['vlow']=fuzz.trimf(pwm.universe,[0,0,60])
pwm['low']=fuzz.trimf(pwm.universe,[0,60,130])
pwm['medium']=fuzz.trimf(pwm.universe,[60,130,200])
pwm['high']=fuzz.trimf(pwm.universe,[130,200,255])
pwm['vhigh']=fuzz.trimf(pwm.universe,[200,200,255])

luat1 = ctrl.Rule(setpoint['low'] & sensor['vlow'],pwm['vlow'])
luat2 = ctrl.Rule(setpoint['low'] & sensor['low'],pwm['vlow'])
luat3 = ctrl.Rule(setpoint['low'] & sensor['medium'],pwm['medium'])
luat4 = ctrl.Rule(setpoint['low'] & sensor['high'],pwm['medium'])
luat5 = ctrl.Rule(setpoint['low'] & sensor['vhigh'],pwm['high'])
luat6 = ctrl.Rule(setpoint['medium'] & sensor['vlow'],pwm['vlow'])
luat7 = ctrl.Rule(setpoint['medium'] & sensor['low'],pwm['medium'])
luat8 = ctrl.Rule(setpoint['medium'] & sensor['medium'],pwm['medium'])
luat9 = ctrl.Rule(setpoint['medium'] & sensor['high'],pwm['high'])
luat10 = ctrl.Rule(setpoint['medium'] & sensor['vhigh'],pwm['high'])
luat11 = ctrl.Rule(setpoint['high'] & sensor['vlow'],pwm['medium'])
luat12= ctrl.Rule(setpoint['high'] & sensor['low'],pwm['medium'])
luat13= ctrl.Rule(setpoint['high'] & sensor['medium'],pwm['high'])
luat14= ctrl.Rule(setpoint['high'] & sensor['high'],pwm['high'])
luat15 = ctrl.Rule(setpoint['high'] & sensor['vhigh'],pwm['vhigh'])

tipping_ctrl = ctrl.ControlSystem([luat1,luat2,luat3,luat4,luat5,luat6,luat7,luat8,luat9,luat10,luat11,luat12,luat13,luat14,luat15])
tipping = ControlSystemSimulation(tipping_ctrl)
tipping.input['setpoint']=20
tipping.input['sensor']=40
tipping.compute()
print(tipping.output['pwm'])
pwm.view(sim = tipping)
```
