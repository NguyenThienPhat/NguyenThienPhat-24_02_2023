```python
from keras.datasets import fashion_mnist
(train_imgers,train_lables),(test_imgers,test_lables)=fashion_mnist.load_data()
print(train_imgers.shape)
print(train_lables.shape)


```

```python
print(test_imgers.shape)
print(test_lables.shape)
```

```python
digit = train_imgers[1610]
import matplotlib.pyplot as plt
plt.imshow(digit)
plt.show
print('so nay la',train_lables[1610])
```

```python
for i in range (25):
  plt.subplot(5,5,i+1)
  plt.tight_layout
  plt.imshow(train_imgers[i],cmap='gray',interpolation='None')
  plt.title('day la la:{}'.format(train_lables[i+1610]))
  plt.xticks([])
  plt.yticks([])
```

```python
from keras.models import Sequential
from keras.layers import Dense
model=Sequential()
model.add(Dense(512,activation='relu',input_shape=(28*28,)))
model.add(Dense(512,activation='relu',input_shape=(28*28,)))
model.add(Dense(512,activation='relu',input_shape=(28*28,)))
model.add(Dense(10,activation='softmax'))
model.summary()
```

```python
train_imgers= train_imgers.reshape((60000,28*28))
train_imgers=train_imgers.astype('float32')/255
train_imgers.shape
```

```python
test_imgers= test_imgers.reshape((10000,28*28))
test_imgers=test_imgers.astype('float32')/255
test_imgers.shape
```

```python
from keras.utils import to_categorical
train_lables=to_categorical(train_lables,10)
test_lables=to_categorical(test_lables,10)
train_lables.shape
```

```python
test_lables.shape
```

```python
model.compile(optimizer='rmsprop',loss='categorical_crossentropy',metrics=['accuracy'])
model.fit(train_imgers,train_lables,epochs=20,batch_size=128)
test_loss,test_acc=model.evaluate(test_imgers,test_lables)
print('Do mat mat:',test_loss)
print('Do chinh xac:',test_acc)
```

```python
model.save('test.h5')
```
